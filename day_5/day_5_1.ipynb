```python
class Dog:
    def __init__(self,name,age):
        self.__name=name
        self.age=age
    def bark(self):
        return "Mooooooof"
    def get_info(self):
        return f"name: {self.__name}\nage: {self.age}"
my_dog=Dog("Buddy",1)
print(my_dog.bark())
print(my_dog.get_info())
print(my_dog.age)
print(my_dog.name)
```

```python
# polymorphism
class Animal:
    def __init__(self,name):
        self.name=name
    def sound(self):
        raise NotImplementedError("Subclass must implement abstract method")
class Dog(Animal):
    def sound(self):
        return "Woof!!"
class Cat(Animal):
    def sound(self):
        return "Meow!!"
dog=Dog("Buddy")
cat=Cat("Whisker")
print(dog.sound())
print(cat.sound())
```

```python
# method overloading
class MathOperation:
    def add(self,a,b,c=0):  # c is optional argument
        return a+b+c
math_ops=MathOperation()
print(math_ops.add(2,3))
print(math_ops.add(2,3,4))
```

```python
import tkinter as tk
from tkinter import messagebox

# Function to update the expression in the entry box
def click_button(item):
    global expression
    expression += str(item)  # Append the clicked button text to the expression
    input_text.set(expression)  # Update the input text variable

# Function to clear the entry box
def clear():
    global expression
    expression = ""  # Clear the expression
    input_text.set("")  # Clear the input text variable

# Function to evaluate the final expression
def evaluate():
    try:
        global expression
        result = str(eval(expression))  # Evaluate the expression and convert result to string
        input_text.set(result)  # Update the input text variable with the result
        expression = ""  # Clear the expression
    except Exception as e:
        messagebox.showerror("Error", "Invalid Input")  # Show error message if evaluation fails
        expression = ""  # Clear the expression

# Create the main window
window = tk.Tk()
window.geometry("400x400")  # Set the window size
window.title("Simple Calculator")  # Set the window title


expression = ""  # Initialize the expression variable
input_text = tk.StringVar()  # Create a StringVar to hold the input text

# Create an entry widget
input_frame = tk.Frame(window)
input_frame.pack(pady=10)  # Add some padding around the frame

input_field = tk.Entry(input_frame, textvariable=input_text, font=('Arial', 18), width=24, bd=5, insertwidth=4, justify='right')
input_field.grid(row=0, column=0)  # Position the entry widget in the frame
input_field.pack(ipady=10)  # Add some padding inside the entry widget

# Create a frame for the buttons
button_frame = tk.Frame(window)
button_frame.pack()  # Pack the frame into the window

# Define button labels
buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    'C', '0', '=', '+',
    # '**', '//', '==', '<',
]

# Create buttons and add them to the frame
row_val = 0
col_val = 0
for button in buttons:
    if button == '=':
        btn = tk.Button(button_frame, text=button, width=10, height=3, bd=5, bg='#ffbf00', command=evaluate)
    elif button == 'C':
        btn = tk.Button(button_frame, text=button, width=10, height=3, bd=5, bg='#ff7f7f', command=clear)
    else:
        btn = tk.Button(button_frame, text=button, width=10, height=3, bd=5,bg="#eaff30", command=lambda b=button: click_button(b))
    btn.grid(row=row_val, column=col_val, padx=5, pady=5)  # Position the button in the grid
    col_val += 1
    if col_val > 3:
        col_val = 0
        row_val += 1

# Run the application
window.mainloop()  # Start the Tkinter event loop
```
